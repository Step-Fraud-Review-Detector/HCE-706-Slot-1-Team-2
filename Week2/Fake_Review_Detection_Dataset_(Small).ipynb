## Data wrangling

### Subtask:
Tokenize the review text and remove stop words.


**Reasoning**:
Tokenize the review text, remove stop words, and display the first 5 rows of the modified dataframe.



```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import string

# Download required resources
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Initialize tools
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    # 1. Lowercase
    text = text.lower()

    # 2. Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # 3. Tokenize
    tokens = word_tokenize(text)

    # 4. Remove stopwords and stem
    tokens = [stemmer.stem(word) for word in tokens if word not in stop_words]

    # 5. Join back to string (optional)
    return ' '.join(tokens)
df['processed_review'] = df['review'].apply(preprocess_text)
df.head()
```

## Data wrangling

### Subtask:
Create separate dataframes for product information, user information, and review information, if feasible.


**Reasoning**:
Create separate dataframes for product, user, and review information based on the available columns.



```python
# Create dataframes for product, user, and review information
products_df = df[['product_id', 'item_title','images']].drop_duplicates().reset_index(drop=True)
categories_df = df[['category_id', 'category']].drop_duplicates().reset_index(drop=True)
reviews_df = df[['rating', 'timestamp', 'label','product_id', 'category_id','review']]

product_avg_rating = df.groupby('product_id')['rating'].mean().reset_index()
product_avg_rating.rename(columns={'rating': 'avg_rating'}, inplace=True)

# Merge into products_df
products_df = products_df.merge(product_avg_rating, on='product_id')

category_avg_rating = df.groupby('category_id')['rating'].mean().reset_index()
category_avg_rating.rename(columns={'rating': 'avg_rating'}, inplace=True)

# Merge into categories_df
categories_df = categories_df.merge(category_avg_rating, on='category_id')


# Display the first few rows and shapes of the new dataframes
display(products_df.head())
print(products_df.shape)
display(reviews_df.head())
print(reviews_df.shape)
display(categories_df.head())
print(categories_df.shape)
```
